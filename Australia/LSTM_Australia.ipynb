```python
#preprocessig the AUD data for LSTM
df_aud = np.array(df_aud).reshape(-1,1)

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

df_aud = scaler.fit_transform(df_aud)

print(df_aud)
```

```python
#Splitting the AUD data into training and test data
aud_train = df_aud[:3008]
aud_test = df_aud[3008:]

print(aud_train.shape)
print(aud_test.shape)
```

```python
def get_data_aud(data_aud, look_back):
    data_aud_x, data_aud_y = [],[]
    for i in range(len(data_aud)-look_back-1):
        data_aud_x.append(data_aud[i:(i+look_back),0])
        data_aud_y.append(data_aud[i+look_back,0])
        
    return np.array(data_aud_x) , np.array(data_aud_y)

look_back = 1

aud_x_train , aud_y_train = get_data_aud(aud_train, look_back)

print(aud_x_train.shape)
print(aud_y_train.shape)
    
```

```python
aud_x_test , aud_y_test = get_data_aud(aud_test, look_back)

print(aud_x_test.shape)
print(aud_y_test.shape)
aud_x_test
```

```python
#Processing train and test sets of AUD for LSTM model

aud_x_train = aud_x_train.reshape(aud_x_train.shape[0],aud_x_train.shape[1], 1)
aud_x_test = aud_x_test.reshape(aud_x_test.shape[0],aud_x_test.shape[1], 1)

print(aud_x_train.shape)
print(aud_x_test.shape) 
```

```python
#Defining the LSTM model for AUD
n_features_aud = aud_x_train.shape[1]
model_aud=Sequential()
model_aud.add(LSTM(100,activation='relu',input_shape=(1,1)))
model_aud.add(Dense(n_features_aud))

#Model summary for AUD
model_aud.summary()
```

```python
#Compiling
model_aud.compile(optimizer='adam', loss = 'mse')

#Training
model_aud.fit(aud_x_train,aud_y_train, epochs = 5, batch_size=1)
```

```python
#Prediction using the trained model for AUD
scaler.scale_

aud_y_pred = model_aud.predict(aud_x_test)
aud_y_pred = scaler.inverse_transform(aud_y_pred)
print(aud_y_pred[:10])
```

```python
def forecast_accuracy(actual, forecast):
    
    mape = np.mean(np.abs(forecast-actual)/np.abs(actual)) #MAPE
    me = np.mean(forecast-actual) #ME
    mae = np.mean(np.abs(forecast-actual)) #MAE
    mpe = np.mean((forecast-actual)/actual) #MPE
    rmse = np.mean((forecast-actual)**2)**.5 #RMSE
    mins = np.amin(np.hstack([forecast[:,None], actual[:,None]]),axis=1)
    maxs = np.amax(np.hstack([forecast[:,None], actual[:,None]]), axis=1)
    minmax = 1 - np.mean(mins/maxs)   # minmax
    accuracy = (1-mape)*100
    return ({'mape':mape, 'me':me, 'mae':mae, 'mpe':mpe, 'rmse':rmse, 'minmax':minmax, 'Accuracy':accuracy})
```
