### Explanatory Data Analysis

```python
# Importing the libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import datetime
import warnings
warnings.filterwarnings('ignore')
```

```python
# Importing the dataset
dataset = pd.read_csv("Foreign_Exchange_Rates.csv")
df = dataset.copy()

print(df.head())
print(df.shape)
print(df.columns)
```

```python
df.head()
```

```python
# Removing the unnusual column name 'Unnamed:0' from the dataset
df = df.drop(columns=df.columns[0])
df.head()
```

```python
# Standardizing the existing column names to "CountryCurrency/USD"
new_col_names = ['Time Series', 'aud', 'brl', 'cad', 'cny', 'dkk', 'eur', 'gbp', 'hkd', 'inr', 'jpy', 'myr', 'mxn', 'nzd', 
                 'nok', 'sgd', 'zar', 'sek', 'chf', 'twd', 'thb']

df.columns = new_col_names

# capitalizing the column names and appending '_USD' to each and every column except the first one
df.columns = map(str.upper, df.columns)
df.rename(columns = lambda x : x + '_USD', inplace=True)
df.rename(columns = {'TIME SERIES_USD' : 'Time Series'}, inplace=True)
```

```python
df.head()
```

```python
# Replacing the 'ND' values with 'NaN'
df.replace(to_replace='ND', value=np.nan, inplace=True)
print("Number of NaN values in a column: ", df.isna().any(axis=1).sum())
```

```python
# Dropping all the NaN values at one go
df = df.dropna().reset_index(drop=True)

#Checking are there any NaN values are present or not
df.isna().sum()
```

```python
df.head()
```

```python
# Extractinng the dataframe to excel file
file_name = 'ExchangeRatesPreprocessed.xlsx'
  
# saving the excel
df.to_excel(file_name)
print('DataFrame is written to Excel File successfully.')
```

### Data Visualization

```python
# data type of each column
df.dtypes
```

```python
# Converting object data type to numeric for each column
for col in df.columns[1:]:
    df[col] = pd.to_numeric(df[col], errors='coerce')  # returns invalid value as NaN
```

```python
# appending 3 more coulmns as month,year and month_year
df['Time Series'] = pd.to_datetime(df['Time Series'])
df['month'] = df['Time Series'].dt.month
df['year'] = df['Time Series'].dt.year
df['month_year'] = df['Time Series'].dt.to_period('M')
```

```python
df.head()
```

```python
print("Shape of Australia:", df['AUD_USD'].count())
print("Shape of European Union:", df['EUR_USD'].count())
print("Shape of United Kingdom:", df['GBP_USD'].count())
```

```python
# Australia
# Taking the average rate of each month, assuming mean represents the entire month
df_groupby_aud = df.groupby('month_year').AUD_USD.mean().reset_index()
df_groupby_aud
```

```python
# European Union
# Taking the average rate of each month, assuming mean represents the entire month
df_groupby_eur = df.groupby('month_year').EUR_USD.mean().reset_index()
df_groupby_eur
```

```python
# United Kingdom
# Taking the average rate of each month, assuming mean represents the entire month
df_groupby_gbp = df.groupby('month_year').GBP_USD.mean().reset_index()
df_groupby_gbp
```

```python
import plotly.express as ps
```

```python
# Creating a dataframe of average rate of each month of all the given countries
groupby_all = df.groupby('month_year').apply(lambda x : pd.Series
                                            ({'AUD_mean': x['AUD_USD'].mean(),'BRL_mean': x['BRL_USD'].mean(),
                                            'CAD_mean': x['CAD_USD'].mean(),'CNY_mean': x['CNY_USD'].mean(),
                                            'DKK_mean': x['DKK_USD'].mean(),'EUR_mean': x['EUR_USD'].mean(),
                                            'GBP_mean': x['GBP_USD'].mean(),'HKD_mean': x['HKD_USD'].mean(),
                                            'INR_mean': x['INR_USD'].mean(),'JPY_mean': x['JPY_USD'].mean(),
                                            'MYR_mean': x['MYR_USD'].mean(),'MXN_mean': x['MXN_USD'].mean(),
                                            'NZD_mean': x['NZD_USD'].mean(),'NOK_mean': x['NOK_USD'].mean(),
                                            'SGD_mean': x['SGD_USD'].mean(),'ZAR_mean': x['ZAR_USD'].mean(),
                                            'SEK_mean': x['SEK_USD'].mean(),'CHF_mean': x['CHF_USD'].mean(),
                                            'TWD_mean': x['TWD_USD'].mean(),'THB_mean': x['THB_USD'].mean()})).reset_index()

groupby_all = pd.DataFrame(groupby_all)

groupby_all['USD_mean'] = [1 for i in range(180)]
groupby_all.head()
```

```python
# to visualize all the countries pattern
y_all = groupby_all[['AUD_mean','BRL_mean','CAD_mean','CNY_mean','DKK_mean','EUR_mean','GBP_mean','HKD_mean','INR_mean',
                     'JPY_mean','NZD_mean','MYR_mean','MXN_mean','SGD_mean','ZAR_mean','USD_mean']]

labels = ['AUD_mean','BRL_mean','CAD_mean','CNY_mean','DKK_mean','EUR_mean','GBP_mean','HKD_mean','INR_mean',
                     'JPY_mean','NZD_mean','MYR_mean','MXN_mean','SGD_mean','ZAR_mean','USD_mean']

x_ticks = list(range(1, 180, 12))
x_ticklabels = [x for x in range(2006, 2021)]

plt.figure(figsize=(20,10))
ax = plt.subplot()

plt.plot(x, y_all)
ax.set_xticks(x_ticks)
ax.set_xticklabels(x_ticklabels)

plt.legend(labels)
plt.title("Exchange Rates: Countries/USD \n", size=20)
plt.xlabel("Year", size=15)
plt.ylabel("Exchange Rates", size=15)
plt.grid(ls="--")
plt.show()
```
